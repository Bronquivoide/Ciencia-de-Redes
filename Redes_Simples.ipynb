### Este notebook tiene como objetivo visualizar tres tipos de redes de estructura simple:

### 1) Red lineal.

### 2) Red circular.

### 3) Red completa.

### El proceso va a consistir en crear una red vacía mediante un objeto de la clase Graph en NetworkX, e ir añadiendo $N$ nodos con $L$ enlaces asociados tal que se genere una red definida por la característica deseada (lineal, circular y completa).

### **Nota: recordemos que una red simple tiene dos características principales: no presenta autoenlaces ni dirección en los enlaces (es decir, es una red no dirigida)**

# 1) Red lineal:


```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt

#Definimos N:
N1=10

#Red vacía:
R_lineal = nx.Graph()

#Nodos y enlaces:
for i in range(N1-1): # Si defino i en N1, por cuestiones de indexación en Python me arrojaría 11 nodos
  R_lineal.add_edge(i,i+1) # El conjunto de enlaces consiste en las parejas (n_i , n_i+1)

#Visualización de la red:
nx.draw(R_lineal,with_labels=True,node_color='purple')
plt.title("Red Lineal",fontsize=25,color="purple")
plt.show()

#Información de la red:
print("Estructura de la red:")
print()
print(f"Los nodos de la red son:\n {R_lineal.nodes}")
print()
print(f"Los enlaces son:\n {R_lineal.edges}")
print()
print(f"Los grados asociados de la red (en formato nodo , su grado) son:\n {R_lineal.degree}")

```

# 2) Red circular:

### Esta clase de red posee un camino tipo ciclo, es decir, el nodo donde el camino inicia es el mismo donde termina. Para esto vamos a implementar un conjunto de enlaces de la forma $\{(n_i,n_{i-1})\}$

```python
#Número de nodos:
N2=15

#Red vacía:
R_circular=nx.Graph()

#Nodos:
for i in range(N2):
  R_circular.add_node(i)

#Almacenamiento de nodos:
lista2= list(R_circular.nodes)

#Enlaces:
for i in range(N2):
  R_circular.add_edge(lista2[i], lista2[i-1])

#Visualización:
nx.draw(R_circular,with_labels=True,node_color='r')
plt.title("Red Circular",fontsize=25,color="r")
plt.show()

#Información de la red:
print("Estructura de la red:")
print()
print(f"Los nodos de la red son:\n {R_circular.nodes}")
print()
print(f"Los enlaces son:\n {R_circular.edges}")
print()
print(f"Los grados asociados de la red (en formato nodo , su grado) son:\n {R_circular.degree}")
```

# 3) Red completa:

### Una red completa tiene como característica que cada nodo está enlazado con todos los demás.


```python
#Red vacía:
N3=15
R_completa=nx.Graph()

#Enlaces:
for i in range(N3):
  for j in range(i):
    R_completa.add_edge(i,j) #Estructura para evitar autoenlaces

#Visualización:
nx.draw(R_completa,with_labels=True,node_color='yellow')
plt.title("Red Completa",fontsize=25,color="orange")
plt.show()

#Información de la red:
print("Estructura de la red:")
print()
print(f"Los nodos de la red son:\n {R_completa.nodes}")
print()
print(f"Los enlaces son:\n {R_completa.edges}")
print()
print(f"Los grados asociados de la red (en formato nodo , su grado) son:\n {R_completa.degree}")
```
