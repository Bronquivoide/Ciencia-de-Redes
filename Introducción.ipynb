### Este notebook tiene como objetivo aplicar nociones básicas de ciencia de redes en Python, sentando las bases para generar y analizar ejemplos básicos y sencillos de redes.

### Usando la paquetería *NetworkX*, vamos a mostrar cómo construir algunas redes de estructura simple.

### Dentro de NetworkX tenemos la clase *Graph()*, que nos genera objetos de esta clasificación.

### Recordemos que una red (o grafo) está definida como un conjunto de $N$ nodos distinguibles $\{n_1,n_2, \dots , n_N\}$ con un conjunto de $L$ enlaces asociados que forman parejas de nodos $\{(n_i , n_j)\}$. Una red está caracterizada por su matriz de adyacencia.

### Vamos a empezar por generar una red que sea un conjunto vacío de nodos y vamos a ir explorando y visualizando su estructura básica añadiendo nodos y enlaces usando NetworkX.

```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt

#Clase Graph:
RED=nx.Graph() #Genera una red sin nodos
print(f"La red está caracterizada por la siguiente estructura: {RED}")
print()

#Visualización de la red:
print("La función nx.draw nos permite visualzar la red, que en este caso no arrojará nada por ser una red vacía:")
nx.draw(RED)
print()

print("La función RED.nodes nos arroja una lista de los nodos, que en este caso es una lista sin elementos:")
RED.nodes
print()

print("La función RED.edges nos arroja la lista de enlaces, la lista de las parejas de nodos que tampoco contiene elementos:")
RED.edges
```

### Para agregar nodos a nuestra red se utiliza la función *add_node()*, la cual recibe como argumento el nombre del nodo.

### Consideremos el conjunto de nodos de la forma $\{n_p \}, \forall p=1,2,3,4,5$. Vamos a añadir los nodos:

```python
#Añadir los nodos:
RED.add_node('n1')
RED.add_node('n2')
RED.add_node('n3')
RED.add_node('n4')
RED.add_node('n5')

#Visualización:
nx.draw(RED, with_labels = True) 
```

### Se agregan los enlaces entre nodos mediante la función *add_edge()*, la cual recibe dos argumentos para enlazar: el nodo de origen y el nodo destino.

### Lo hacemos de la siguiente manera y volvemos a mostrar sus propiedades que ya no son conjuntos vacíos:

```python
#Agregamos los enlaces:
RED.add_edge('n1','n2')
RED.add_edge('n1','n3')
RED.add_edge('n1','n4')
RED.add_edge('n3','n5')

#Visualización:
nx.draw(RED, with_labels = True)

#Características de la red:
print(f"Los nodos de la red son:\n {RED.nodes}")
print()
print(f"Los enlaces son:\n {RED.edges}")
print()
print(f"Los grados asociados a los nodos son:\n {RED.degree}")
```
