### La matriz de adyacencia asociada a una red es una propiedad matemática que consiste en un arreglo de datos con la información que describe la estructura de la red.

### Para una red de $N$ nodos, se define una matriz $A \in M_{N \times N} (\mathbb{N}^+ _0)$ (en donde $\mathbb{N}^+ _0$ representa al conjunto $\mathbb{N}^+ \cup \{0\}$) cuyas entradas son de la siguiente forma:

<font size="5"> $ A_{ij} = 
\begin{cases}
m & \text{, si hay $m \in \mathbb{N}^+$ enlaces del nodo $j$ a los nodos $\{i \neq j\}$}  \\
0 & \text{, si no hay enlace}
\end{cases} $</font> 

### En este repositorio se va a tomar la convención de que las columnas de la matriz (entradas de índice $j$) son el nodo de origen del enlace, mientras que los renglones (entradas de índice $i$) son el nodo destino del enlace. Esto sobretodo toma más relevancia en redes dirigidas.

### Por ejemplo, consideremos la siguiente red:



```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

#Red vacía:
red=nx.Graph()

#Nodos de la red:
red.add_nodes_from([1, 2, 3, 4])

#Enlaces en los nodos:
red.add_edges_from([(1, 2), (1, 3), (1, 4), (3, 4)])
nx.draw(red, with_labels=True, node_color='g', font_size=20, node_size=1000)
plt.show()
```

### La matriz de adyacencia de nuestra red sería una matriz $A$ de $4\times 4$. Considerando que no definimos autoenlaces en la red, tenemos que:

<font size="5"> $A = 
\begin{pmatrix}
A_{11} & A_{12} & A_{13} & A_{14} \\
A_{21} & A_{22} & A_{23} & A_{24} \\
A_{31} & A_{32} & A_{33} & A_{34} \\
A_{41} & A_{42} & A_{43} & A_{44}
\end{pmatrix}= \begin{pmatrix}
0 & 1 & 1 & 1 \\
1 & 0 & 0 & 0 \\
1 & 0 & 0 & 1 \\
1 & 0 & 1 & 0
\end{pmatrix}  $</font> 

### Notemos que se tienen las siguientes propiedades:

### 1) La condición de no autoenlaces se traduce analíticamente como $A_{ii}=0$, y se refleja que $tr(A)=0$ puesto que todos los elementos en la diagonal son nulos. Es decir, el nodo de enlace origen no puede ser el mismo de enlace destino.

### 2) Notemos que la red no es dirijida, es decir, no se presenta una dirección en los enlaces. El enlace que sale del nodo $j$ y llega al $i$, representa al mismo que sale de $i$ y llega a $j$. Esto se traduce como una simetría $A_{ij}=A_{ji}$

### 3) <font size="5"> $ A_{ij} = 
\begin{cases}
1 & \text{, si hay enlace del nodo $j$ al nodo $i$}  \\
0 & \text{, si no hay enlace}
\end{cases} $</font> 


### Y parece un buen momento para recalcar que se generalizó la notación $A \in M_{N \times N} (\mathbb{N}_0)$ en vez de solo decir $A \in M_{N \times N} ([0,1] \subset \mathbb{Z}^+ _0)$, puesto que puede haber redes múltiples tal que a un nodo se le asocien $(m>1) \in \mathbb{N}_0$ enlaces. En ese caso la matriz de adyacencia ya no consistiría en valores binarios en $[0,1]$ como en el ejemplo.

### Podemos trabajar las matrices de adyacencias asociadas a objetos de clase Graph en NetworkX con la biblioteca NumPy. Podemos generar una red a partir de una matriz usando la función *nx.from_numpy_array( )*, y podemos asignar una matriz de adyacencia a una red usando *nx.to_numpy_array()*.

# Ejercicio 1:

### Generar una red a partir de la matriz ejemplificada con la función nx.from_numpy_array():

<font size="5"> $ A=\begin{pmatrix}
0 & 1 & 1 & 1 \\
1 & 0 & 0 & 0 \\
1 & 0 & 0 & 1 \\
1 & 0 & 1 & 0
\end{pmatrix} $
</font>


```python
#Generar la matriz:
matriz=[[0,1,1,1], [1,0,0,0], [1,0,0,1], [1,0,1,0]]

#Hacer a la matriz un arreglo NumPy:
A=np.array(matriz)
print(f"La matriz es:\n {A}")
print()
print(f"Es de orden: {A.shape}")
print()
print(f"Y es de la clase: {type(A)}")

#Generar e imprimir la red asociada a partir de la matriz de adyacencia A:
red=nx.from_numpy_array(A)
nx.draw_circular(red,with_labels = True) #En otras posiciones
plt.title("Red asociada a la matriz A",fontsize=20)
plt.show()
```

### Podemos representar a la matriz de entradas binarias mediante una imagen, pues como justamente tenemos un arreglo de valores binarios, podemos asociar estos a un conjunto de intensidades en una imagen binaria:

```python
plt.imshow(A, cmap="gray",)
plt.title("Representación de la matriz de adyacencia $A$ \n mediante una imagen binaria", fontsize=20)
plt.show()
```

# Ejercicio 2:

### Para aumentar la complejidad, vamos a retomar la red completa que habíamos creado en el notebook de redes simples y vamos a generar su matriz de adyacencia, $A_2$, con la función nx.to_numpy_array():

```python
#Red vacía:
N3=15
completa=nx.Graph()

#Enlaces:
for i in range(N3):
  for j in range(i):
    completa.add_edge(i,j) #Estructura para evitar autoenlaces

#Visualización:
nx.draw(completa,with_labels=True,node_color='yellow')
plt.title("Red Completa",fontsize=25,color="orange")
plt.show()

#Matriz de adyacencia A2 a partir de la red:
A2=nx.to_numpy_array(completa)
print(f"La matriz es:\n {A2}")
print()
print(f"Es de orden: {A2.shape}")
print()
print(f"Y es de la clase: {type(A2)}")

#Imagen binaria de A2:
plt.imshow(A2, cmap="gray",)
plt.title("Representación de la matriz de adyacencia $A_2$ \n mediante una imagen binaria", fontsize=20)
plt.show()
```
